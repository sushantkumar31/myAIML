Topic 1: Linear Algebra for ML

```python
import numpy as np

# ── 1. Vectors ──────────────────────────────────────────────
# Think of these as two students' scores: [Math, Physics, Chemistry]
student_a = np.array([85, 90, 78])
student_b = np.array([70, 88, 95])

print("=== VECTORS ===")
print(f"Student A scores: {student_a}")
print(f"Student B scores: {student_b}")

# Vector addition (combine features)
print(f"Sum: {student_a + student_b}")

# Scalar multiplication (scale all scores)
print(f"Student A scaled by 0.1: {student_a * 0.1}")

# ── 2. Dot Product ──────────────────────────────────────────
# Weights represent how much each subject matters
weights = np.array([0.5, 0.3, 0.2])  # Math matters most

# This is literally what a neuron computes!
weighted_score_a = np.dot(student_a, weights)
weighted_score_b = np.dot(student_b, weights)

print("\n=== DOT PRODUCT (neuron simulation) ===")
print(f"Student A weighted score: {weighted_score_a}")
print(f"Student B weighted score: {weighted_score_b}")

# ── 3. Matrix — full dataset ────────────────────────────────
# 4 students, 3 features each
X = np.array([
    [85, 90, 78],
    [70, 88, 95],
    [60, 72, 80],
    [95, 85, 88]
])

print("\n=== MATRIX (dataset) ===")
print(f"Shape: {X.shape}  → (samples, features)")
print(X)

# ── 4. Matrix Multiplication ─────────────────────────────────
# Weight matrix: 3 input features → 2 output neurons
W = np.array([
    [0.5, 0.2],   # weights from feature 1
    [0.3, 0.4],   # weights from feature 2
    [0.2, 0.4],   # weights from feature 3
])

# One forward pass through a neural network layer!
output = X @ W   # (4×3) @ (3×2) → (4×2)

print("\n=== MATRIX MULTIPLY (neural network forward pass) ===")
print(f"Input shape:  {X.shape}")
print(f"Weight shape: {W.shape}")
print(f"Output shape: {output.shape}  → 4 students, 2 neuron outputs")
print(output)

# ── 5. Transpose ─────────────────────────────────────────────
print("\n=== TRANSPOSE ===")
print(f"W shape: {W.shape}  →  W.T shape: {W.T.shape}")

# ── 6. Eigenvalues & Eigenvectors ────────────────────────────
# Compute on the covariance matrix (used in PCA!)
cov_matrix = np.cov(X.T)   # shape (3, 3)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\n=== EIGENVALUES & EIGENVECTORS ===")
print(f"Covariance matrix shape: {cov_matrix.shape}")
print(f"Eigenvalues:  {np.round(eigenvalues, 2)}")
print("Eigenvectors (columns):")
print(np.round(eigenvectors, 2))
print("\n→ The largest eigenvalue tells you the direction of MOST variance in your data")
print(f"→ Most important direction: eigenvalue = {np.round(max(eigenvalues), 2)}")
```

 Topic 2: Calculus for ML
 

```python
import numpy as np


# ── 1. Numerical Derivative ──────────────────────────────────
# We approximate f'(x) = [f(x+h) - f(x)] / h
def f(x):
    return x ** 2 + 3 * x + 2   # a simple loss-like curve

def numerical_derivative(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

print("=== DERIVATIVES ===")
for x in [-3, -1.5, 0, 2]:
    slope = numerical_derivative(f, x)
    print(f"  f'({x:>4}) = {slope:.4f}")

print("\n→ At x=-1.5 slope≈0, that's the minimum of f(x)=x²+3x+2")

# ── 2. Chain Rule — manual example ──────────────────────────
# Suppose: x → g(x) = x² → f(g) = 3g + 1
# dy/dx = f'(g) * g'(x) = 3 * 2x = 6x
print("\n=== CHAIN RULE ===")
x = 4
g = x ** 2          # inner function
dg_dx = 2 * x       # g'(x)
df_dg = 3           # f'(g) = 3 (since f(g) = 3g + 1)
dy_dx = df_dg * dg_dx
print(f"  x={x},  g(x)={g},  dg/dx={dg_dx},  df/dg={df_dg}")
print(f"  Chain rule → dy/dx = {df_dg} × {dg_dx} = {dy_dx}")

# ── 3. Gradient Descent from Scratch ────────────────────────
# Goal: find the minimum of Loss(w) = (w - 3)²
# True minimum is at w = 3
# Derivative: dLoss/dw = 2*(w - 3)

print("\n=== GRADIENT DESCENT ===")

def loss(w):
    return (w - 3) ** 2

def grad_loss(w):
    return 2 * (w - 3)

w = 10.0          # start far from the answer
lr = 0.1          # learning rate
history = []

for step in range(20):
    L = loss(w)
    g = grad_loss(w)
    history.append((w, L))

    if step < 5 or step == 19:
        print(f"  Step {step+1:>2}: w={w:.4f}  loss={L:.4f}  grad={g:.4f}")

    w = w - lr * g   # gradient descent update

print(f"\n  → Converged to w ≈ {w:.6f}  (true minimum = 3.0)")

# ── 4. Real ML scenario: Linear Regression with GD ──────────
print("\n=== LINEAR REGRESSION VIA GRADIENT DESCENT ===")
np.random.seed(42)

# True relationship: y = 2x + 1 (we want model to learn this)
X = np.random.randn(50)
y = 2 * X + 1 + np.random.randn(50) * 0.3   # add noise

# Initialize parameters
w = 0.0   # weight (should learn ≈ 2)
b = 0.0   # bias   (should learn ≈ 1)
lr = 0.05
n = len(X)

for epoch in range(200):
    # Forward pass
    y_pred = w * X + b

    # Loss: Mean Squared Error
    loss_val = np.mean((y_pred - y) ** 2)

    # Gradients (calculus applied to MSE)
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)

    # Update
    w -= lr * dw
    b -= lr * db

    if epoch % 50 == 0:
        print(f"  Epoch {epoch:>3}: loss={loss_val:.4f}  w={w:.4f}  b={b:.4f}")

print(f"\n  → Learned: y = {w:.3f}x + {b:.3f}")
print(f"  → Truth:   y = 2.000x + 1.000")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Loss function
def loss(w):
    return w**4 - 4*w**2 + w

# Gradient
def grad(w):
    return 4*w**3 - 8*w + 1


def gradient_descent(start_w, lr=0.01, steps=100):
    w = start_w

    w_history = []
    loss_history = []

    for step in range(steps):
        L = loss(w)

        w_history.append(w)
        loss_history.append(L)

        if step % 10 == 0:
            print(
                f"Step {step:3d}: "
                f"w={w:8.4f} "
                f"loss={L:10.4f}"
            )

        w = w - lr * grad(w)

    return w, w_history, loss_history


print("=== Start from w = 3 ===")
final_w_pos, w_hist_pos, loss_hist_pos = gradient_descent(3.0)

print(f"\nFinal w = {final_w_pos:.6f}")

print("\n=== Start from w = -3 ===")
final_w_neg, w_hist_neg, loss_hist_neg = gradient_descent(-3.0)

print(f"\nFinal w = {final_w_neg:.6f}")

# Plot
plt.figure(figsize=(8,5))
plt.plot(loss_hist_pos, label="start w=3")
plt.plot(loss_hist_neg, label="start w=-3")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Gradient Descent on L(w)")
plt.legend()
plt.grid(True)
plt.show()
```
