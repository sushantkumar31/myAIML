Topic 1: Linear Algebra for ML

```python
import numpy as np

# ── 1. Vectors ──────────────────────────────────────────────
# Think of these as two students' scores: [Math, Physics, Chemistry]
student_a = np.array([85, 90, 78])
student_b = np.array([70, 88, 95])

print("=== VECTORS ===")
print(f"Student A scores: {student_a}")
print(f"Student B scores: {student_b}")

# Vector addition (combine features)
print(f"Sum: {student_a + student_b}")

# Scalar multiplication (scale all scores)
print(f"Student A scaled by 0.1: {student_a * 0.1}")

# ── 2. Dot Product ──────────────────────────────────────────
# Weights represent how much each subject matters
weights = np.array([0.5, 0.3, 0.2])  # Math matters most

# This is literally what a neuron computes!
weighted_score_a = np.dot(student_a, weights)
weighted_score_b = np.dot(student_b, weights)

print("\n=== DOT PRODUCT (neuron simulation) ===")
print(f"Student A weighted score: {weighted_score_a}")
print(f"Student B weighted score: {weighted_score_b}")

# ── 3. Matrix — full dataset ────────────────────────────────
# 4 students, 3 features each
X = np.array([
    [85, 90, 78],
    [70, 88, 95],
    [60, 72, 80],
    [95, 85, 88]
])

print("\n=== MATRIX (dataset) ===")
print(f"Shape: {X.shape}  → (samples, features)")
print(X)

# ── 4. Matrix Multiplication ─────────────────────────────────
# Weight matrix: 3 input features → 2 output neurons
W = np.array([
    [0.5, 0.2],   # weights from feature 1
    [0.3, 0.4],   # weights from feature 2
    [0.2, 0.4],   # weights from feature 3
])

# One forward pass through a neural network layer!
output = X @ W   # (4×3) @ (3×2) → (4×2)

print("\n=== MATRIX MULTIPLY (neural network forward pass) ===")
print(f"Input shape:  {X.shape}")
print(f"Weight shape: {W.shape}")
print(f"Output shape: {output.shape}  → 4 students, 2 neuron outputs")
print(output)

# ── 5. Transpose ─────────────────────────────────────────────
print("\n=== TRANSPOSE ===")
print(f"W shape: {W.shape}  →  W.T shape: {W.T.shape}")

# ── 6. Eigenvalues & Eigenvectors ────────────────────────────
# Compute on the covariance matrix (used in PCA!)
cov_matrix = np.cov(X.T)   # shape (3, 3)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\n=== EIGENVALUES & EIGENVECTORS ===")
print(f"Covariance matrix shape: {cov_matrix.shape}")
print(f"Eigenvalues:  {np.round(eigenvalues, 2)}")
print("Eigenvectors (columns):")
print(np.round(eigenvectors, 2))
print("\n→ The largest eigenvalue tells you the direction of MOST variance in your data")
print(f"→ Most important direction: eigenvalue = {np.round(max(eigenvalues), 2)}")
```

 Topic 2: Calculus for ML
 

```python
import numpy as np


# ── 1. Numerical Derivative ──────────────────────────────────
# We approximate f'(x) = [f(x+h) - f(x)] / h
def f(x):
    return x ** 2 + 3 * x + 2   # a simple loss-like curve

def numerical_derivative(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

print("=== DERIVATIVES ===")
for x in [-3, -1.5, 0, 2]:
    slope = numerical_derivative(f, x)
    print(f"  f'({x:>4}) = {slope:.4f}")

print("\n→ At x=-1.5 slope≈0, that's the minimum of f(x)=x²+3x+2")

# ── 2. Chain Rule — manual example ──────────────────────────
# Suppose: x → g(x) = x² → f(g) = 3g + 1
# dy/dx = f'(g) * g'(x) = 3 * 2x = 6x
print("\n=== CHAIN RULE ===")
x = 4
g = x ** 2          # inner function
dg_dx = 2 * x       # g'(x)
df_dg = 3           # f'(g) = 3 (since f(g) = 3g + 1)
dy_dx = df_dg * dg_dx
print(f"  x={x},  g(x)={g},  dg/dx={dg_dx},  df/dg={df_dg}")
print(f"  Chain rule → dy/dx = {df_dg} × {dg_dx} = {dy_dx}")

# ── 3. Gradient Descent from Scratch ────────────────────────
# Goal: find the minimum of Loss(w) = (w - 3)²
# True minimum is at w = 3
# Derivative: dLoss/dw = 2*(w - 3)

print("\n=== GRADIENT DESCENT ===")

def loss(w):
    return (w - 3) ** 2

def grad_loss(w):
    return 2 * (w - 3)

w = 10.0          # start far from the answer
lr = 0.1          # learning rate
history = []

for step in range(20):
    L = loss(w)
    g = grad_loss(w)
    history.append((w, L))

    if step < 5 or step == 19:
        print(f"  Step {step+1:>2}: w={w:.4f}  loss={L:.4f}  grad={g:.4f}")

    w = w - lr * g   # gradient descent update

print(f"\n  → Converged to w ≈ {w:.6f}  (true minimum = 3.0)")

# ── 4. Real ML scenario: Linear Regression with GD ──────────
print("\n=== LINEAR REGRESSION VIA GRADIENT DESCENT ===")
np.random.seed(42)

# True relationship: y = 2x + 1 (we want model to learn this)
X = np.random.randn(50)
y = 2 * X + 1 + np.random.randn(50) * 0.3   # add noise

# Initialize parameters
w = 0.0   # weight (should learn ≈ 2)
b = 0.0   # bias   (should learn ≈ 1)
lr = 0.05
n = len(X)

for epoch in range(200):
    # Forward pass
    y_pred = w * X + b

    # Loss: Mean Squared Error
    loss_val = np.mean((y_pred - y) ** 2)

    # Gradients (calculus applied to MSE)
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)

    # Update
    w -= lr * dw
    b -= lr * db

    if epoch % 50 == 0:
        print(f"  Epoch {epoch:>3}: loss={loss_val:.4f}  w={w:.4f}  b={b:.4f}")

print(f"\n  → Learned: y = {w:.3f}x + {b:.3f}")
print(f"  → Truth:   y = 2.000x + 1.000")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Loss function
def loss(w):
    return w**4 - 4*w**2 + w

# Gradient
def grad(w):
    return 4*w**3 - 8*w + 1


def gradient_descent(start_w, lr=0.01, steps=100):
    w = start_w

    w_history = []
    loss_history = []

    for step in range(steps):
        L = loss(w)

        w_history.append(w)
        loss_history.append(L)

        if step % 10 == 0:
            print(
                f"Step {step:3d}: "
                f"w={w:8.4f} "
                f"loss={L:10.4f}"
            )

        w = w - lr * grad(w)

    return w, w_history, loss_history


print("=== Start from w = 3 ===")
final_w_pos, w_hist_pos, loss_hist_pos = gradient_descent(3.0)

print(f"\nFinal w = {final_w_pos:.6f}")

print("\n=== Start from w = -3 ===")
final_w_neg, w_hist_neg, loss_hist_neg = gradient_descent(-3.0)

print(f"\nFinal w = {final_w_neg:.6f}")

# Plot
plt.figure(figsize=(8,5))
plt.plot(loss_hist_pos, label="start w=3")
plt.plot(loss_hist_neg, label="start w=-3")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Gradient Descent on L(w)")
plt.legend()
plt.grid(True)
plt.show()
```

Topic 3: Probability & Statistics for ML


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# ── 1. Distributions ─────────────────────────────────────────
print("=== DISTRIBUTIONS ===")

uniform = np.random.uniform(0, 1, 1000)
gaussian = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1
bernoulli = np.random.binomial(n=1, p=0.3, size=1000)  # P(success)=0.3

print(f"Uniform  → mean: {uniform.mean():.3f}  std: {uniform.std():.3f}")
print(f"Gaussian → mean: {gaussian.mean():.3f}  std: {gaussian.std():.3f}")
print(f"Bernoulli→ mean: {bernoulli.mean():.3f}  (≈ p=0.3)")

# ── 2. Mean, Variance, Std ───────────────────────────────────
print("\n=== MEAN / VARIANCE / STD ===")
data = np.array([10, 20, 20, 30, 40, 40, 40, 50])

mean = np.mean(data)
var  = np.var(data)
std  = np.std(data)

print(f"Data: {data}")
print(f"Mean:     {mean}")
print(f"Variance: {var:.2f}")
print(f"Std Dev:  {std:.2f}")

# Manual verification
manual_mean = sum(data) / len(data)
manual_var  = sum((x - manual_mean)**2 for x in data) / len(data)
print(f"\nManual mean: {manual_mean}  ✓" if manual_mean == mean else "mismatch")

# ── 3. Bayes' Theorem ────────────────────────────────────────
print("\n=== BAYES' THEOREM (spam filter) ===")

p_spam         = 0.3    # prior: 30% of emails are spam
p_free_given_spam   = 0.8    # likelihood
p_free_given_ham    = 0.1    # only 10% of normal emails say FREE
p_free         = p_free_given_spam * p_spam + p_free_given_ham * (1 - p_spam)

p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P(spam)              = {p_spam}")
print(f"P('FREE' | spam)     = {p_free_given_spam}")
print(f"P('FREE')            = {p_free:.3f}")
print(f"P(spam | 'FREE')     = {p_spam_given_free:.3f}")
print(f"→ Seeing 'FREE' updated belief from 30% → {p_spam_given_free*100:.1f}%")

# ── 4. Expectation ───────────────────────────────────────────
print("\n=== EXPECTATION ===")
# Fair die: values 1-6, each with prob 1/6
values = np.array([1, 2, 3, 4, 5, 6])
probs  = np.array([1/6] * 6)
expectation = np.sum(values * probs)
print(f"E[die roll] = {expectation:.3f}  (expected: 3.5)")

# ── 5. Covariance & Correlation ──────────────────────────────
print("\n=== COVARIANCE & CORRELATION ===")
study_hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
test_scores = np.array([50, 55, 60, 65, 70, 75, 85, 95])
sleep_hours = np.array([8, 7, 7, 6, 5, 4, 4, 3])  # inverse relationship

cov_study_score = np.cov(study_hours, test_scores)[0, 1]
corr_study_score = np.corrcoef(study_hours, test_scores)[0, 1]
corr_sleep_score = np.corrcoef(sleep_hours, test_scores)[0, 1]

print(f"Cov(study, score)  = {cov_study_score:.2f}  (positive → move together)")
print(f"Corr(study, score) = {corr_study_score:.3f}  (strong positive)")
print(f"Corr(sleep, score) = {corr_sleep_score:.3f}  (negative relationship)")

# ── 6. Central Limit Theorem ─────────────────────────────────
print("\n=== CENTRAL LIMIT THEOREM ===")
# Original distribution: highly skewed (exponential)
population = np.random.exponential(scale=2, size=100000)
print(f"Population mean: {population.mean():.3f}  skew (not Gaussian)")

# Take 1000 samples of size 30, compute each mean
sample_means = [np.random.choice(population, 30).mean() for _ in range(1000)]
sample_means = np.array(sample_means)
print(f"Sample means → mean: {sample_means.mean():.3f}  std: {sample_means.std():.3f}")
print(f"→ Sample means follow a Gaussian even though population is exponential!")
```

Topic 4: Information Theory for ML


```python
import numpy as np
import matplotlib.pyplot as plt

# ── 1. Self-Information ──────────────────────────────────────
print("=== SELF-INFORMATION ===")

def self_information(p):
    return -np.log2(p)

events = {"Sun rises":0.9999, "Rain in Hyderabad (monsoon)":0.7,
          "Roll a 6":1/6, "Win lottery":1e-6}

for event, prob in events.items():
    print(f"  P={prob:.6f} → surprise={self_information(prob):.3f} bits  [{event}]")

# ── 2. Entropy ───────────────────────────────────────────────
print("\n=== ENTROPY ===")

def entropy(probs):
    probs = np.array(probs)
    # Avoid log(0) — 0*log(0) is defined as 0 in information theory
    probs = probs[probs > 0]
    return -np.sum(probs * np.log2(probs))

distributions = {
    "Certain      [1.0, 0.0]": [1.0, 0.0],
    "Skewed       [0.9, 0.1]": [0.9, 0.1],
    "Moderate     [0.7, 0.3]": [0.7, 0.3],
    "Fair coin    [0.5, 0.5]": [0.5, 0.5],
    "3-class even [.33,.33,.33]": [1/3, 1/3, 1/3],
}

for name, dist in distributions.items():
    print(f"  H={entropy(dist):.4f} bits  ← {name}")

# ── 3. Cross-Entropy Loss ─────────────────────────────────────
print("\n=== CROSS-ENTROPY LOSS ===")

def cross_entropy(p_true, q_pred):
    q_pred = np.clip(q_pred, 1e-12, 1.0)   # avoid log(0)
    return -np.sum(p_true * np.log(q_pred))  # natural log (nats), standard in ML

# True label: class 0 (cat) in a 3-class problem
p_true = np.array([1.0, 0.0, 0.0])

predictions = {
    "Confident & correct  [0.9, 0.05, 0.05]": [0.90, 0.05, 0.05],
    "Uncertain            [0.4, 0.3,  0.3 ]": [0.40, 0.30, 0.30],
    "Confident & wrong    [0.05,0.9, 0.05 ]": [0.05, 0.90, 0.05],
}

for desc, q in predictions.items():
    loss = cross_entropy(p_true, np.array(q))
    print(f"  Loss={loss:.4f}  ← {desc}")

print("\n→ Confident & wrong = highest loss. This drives the model to be calibrated.")

# ── 4. KL Divergence ─────────────────────────────────────────
print("\n=== KL DIVERGENCE ===")

def kl_divergence(p, q):
    p = np.array(p, dtype=float)
    q = np.array(q, dtype=float)
    q = np.clip(q, 1e-12, 1.0)
    mask = p > 0
    return np.sum(p[mask] * np.log(p[mask] / q[mask]))

p_true = np.array([0.5, 0.3, 0.2])      # true distribution
q_close = np.array([0.45, 0.32, 0.23])  # close model
q_far   = np.array([0.1,  0.1,  0.8])   # far-off model

print(f"  KL(p || q_close) = {kl_divergence(p_true, q_close):.4f}  (similar → low)")
print(f"  KL(p || q_far)   = {kl_divergence(p_true, q_far):.4f}  (different → high)")
print(f"  KL(p || p)       = {kl_divergence(p_true, p_true):.4f}  (identical → zero)")

# ── 5. Entropy across probability range ──────────────────────
print("\n=== ENTROPY CURVE ===")
p_values = np.linspace(0.001, 0.999, 200)
entropies = [-p*np.log2(p) - (1-p)*np.log2(1-p) for p in p_values]

plt.figure(figsize=(7, 4))
plt.plot(p_values, entropies, color='steelblue', linewidth=2)
plt.xlabel("P(heads)")
plt.ylabel("Entropy (bits)")
plt.title("Binary Entropy — Maximum Uncertainty at p=0.5")
plt.axvline(0.5, color='red', linestyle='--', label='max entropy at p=0.5')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# ============================================================
# DATA
# ============================================================

y_true = np.array([
    [1, 0, 0],   # cat
    [0, 1, 0],   # dog
    [0, 0, 1],   # bird
    [1, 0, 0],   # cat
])

y_pred = np.array([
    [0.80, 0.15, 0.05],   # pretty sure cat
    [0.30, 0.60, 0.10],   # leans dog
    [0.10, 0.10, 0.80],   # pretty sure bird
    [0.40, 0.40, 0.20],   # confused on cat
])

# ============================================================
# FUNCTIONS
# ============================================================

def cross_entropy_per_sample(y_true, y_pred):
    """
    Returns cross-entropy for each sample.
    """
    y_pred = np.clip(y_pred, 1e-12, 1.0)
    return -np.sum(y_true * np.log(y_pred), axis=1)


def entropy(p):
    """
    Entropy of a probability distribution.
    """
    p = np.array(p, dtype=float)
    p = p[p > 0]
    return -np.sum(p * np.log(p))


def kl_divergence(p, q):
    """
    KL(P || Q)
    """
    p = np.array(p, dtype=float)
    q = np.array(q, dtype=float)

    q = np.clip(q, 1e-12, 1.0)

    mask = p > 0

    return np.sum(
        p[mask] * np.log(p[mask] / q[mask])
    )


def evaluate_predictions(y_true, y_pred):
    """
    Returns per-sample CE and mean CE.
    """
    losses = cross_entropy_per_sample(y_true, y_pred)
    mean_loss = np.mean(losses)

    return losses, mean_loss


# ============================================================
# TODO 1
# PER-SAMPLE CROSS-ENTROPY
# ============================================================

print("=== TODO 1: PER-SAMPLE CROSS-ENTROPY ===")

losses = cross_entropy_per_sample(y_true, y_pred)

for i, loss in enumerate(losses):
    print(f"Sample {i+1}: loss = {loss:.4f}")

highest_idx = np.argmax(losses)

print(
    f"\nHighest loss: Sample {highest_idx+1} "
    f"(loss={losses[highest_idx]:.4f})"
)

# ============================================================
# TODO 2
# MEAN CROSS-ENTROPY
# ============================================================

print("\n=== TODO 2: MEAN CROSS-ENTROPY ===")

mean_loss = np.mean(losses)

print(f"Mean loss = {mean_loss:.4f}")

# ============================================================
# TODO 3
# ENTROPY OF TRUE LABELS
# ============================================================

print("\n=== TODO 3: ENTROPY OF TRUE LABELS ===")

for i, row in enumerate(y_true):
    h = entropy(row)
    print(f"Sample {i+1}: entropy = {h:.4f}")

print(
    "\nObservation:"
    "\nAll entropies are 0 because one-hot labels "
    "contain no uncertainty."
)

# ============================================================
# TODO 4
# KL DIVERGENCE
# ============================================================

print("\n=== TODO 4: KL DIVERGENCE ===")

kl_sample1 = kl_divergence(y_true[0], y_pred[0])
kl_sample4 = kl_divergence(y_true[3], y_pred[3])

print(f"KL(y_true[0] || y_pred[0]) = {kl_sample1:.4f}")
print(f"KL(y_true[3] || y_pred[3]) = {kl_sample4:.4f}")

if kl_sample4 > kl_sample1:
    print("\nSample 4 has higher KL divergence.")
    print("This matches the cross-entropy result.")

# ============================================================
# STRETCH GOAL
# GENERAL FUNCTION
# ============================================================

print("\n=== STRETCH GOAL ===")

losses, mean_loss = evaluate_predictions(
    y_true,
    y_pred
)

print(f"Original model mean loss = {mean_loss:.4f}")

# ------------------------------------------------------------
# PERFECT PREDICTIONS
# ------------------------------------------------------------

perfect_pred = np.array([
    [0.999, 0.0005, 0.0005],
    [0.0005, 0.999, 0.0005],
    [0.0005, 0.0005, 0.999],
    [0.999, 0.0005, 0.0005],
])

perfect_losses, perfect_mean = evaluate_predictions(
    y_true,
    perfect_pred
)

print(f"\nPerfect model mean loss = {perfect_mean:.6f}")

# ------------------------------------------------------------
# TERRIBLE PREDICTIONS
# ------------------------------------------------------------

terrible_pred = np.array([
    [0.05, 0.90, 0.05],
    [0.05, 0.05, 0.90],
    [0.90, 0.05, 0.05],
    [0.05, 0.90, 0.05],
])

terrible_losses, terrible_mean = evaluate_predictions(
    y_true,
    terrible_pred
)

print(f"Terrible model mean loss = {terrible_mean:.4f}")

# ============================================================
# SUMMARY
# ============================================================

print("\n=== SUMMARY ===")
print(f"Original mean loss : {mean_loss:.4f}")
print(f"Perfect mean loss  : {perfect_mean:.6f}")
print(f"Terrible mean loss : {terrible_mean:.4f}")

print("\nCross-entropy is smallest when the model")
print("assigns high probability to the true class.")
```
